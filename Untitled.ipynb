```python
## Show all playerids and salaries with a salary in the year 1985 above 500k.

## set up query
tbl = 'salaries'
query = 'SELECT * FROM ' + tbl

## execute
salaries = pd.read_sql_query(query, cnx)
print salaries.head()

## subselect
salaries_filtered = salaries[(salaries.yearid == 1985) & (salaries.salary >= 500000)]
print salaries_filtered.min()

## finish
answer = salaries_filtered[['playerid','salary']]
answer.head()
```

```python
# Show the team for each year that had a rank of 1.

## set up query
tbl = 'teams'
query = 'SELECT * FROM ' + tbl

## execute
teams = pd.read_sql_query(query, cnx)
teams = teams[['yearid', 'rank', 'name']]

## subselect
teams_filtered = teams[teams['rank'] == 1]
print teams_filtered.head()
```

```python
# How many schools are in schoolstate of CT?

## set up query
tbl = 'schools'
query = 'SELECT * FROM ' + tbl

## execute
schools = pd.read_sql_query(query, cnx)
schools.head()

## subselect 
schools_filtered = schools[schools.schoolstate == 'CT']
schools_filtered.schoolid.count()
```

```python
# How many schools are there in each state?
schools_grouped = schools.groupby('schoolstate')

## aggregate
schools_grouped.schoolid.count().head(10)
```

```python
# Find all of the salaries of shortstops (fieldings, pos) for the year 2012.

## set up query
tbl = 'fielding'
query = 'SELECT * FROM ' + tbl

## execute
fielding = pd.read_sql_query(query, cnx)
print fielding.head()

## subselect
ss2012 = fielding[(fielding.pos == 'SS') & (fielding.yearid == 2012)][['playerid','yearid']]
ss2012.head()

## join
ss2012_salary = ss2012.merge(salaries, on=('playerid','yearid'))
ss2012_salary.head()
```

```python
# What is the first and last year played for each player?

## group by
fielding_grouped = fielding.groupby('playerid')

## you can return multiple aggregates on the same column using .agg()
fielding_grouped.agg({'yearid':[min, max]}).head()
```

```python
# Who has played the most all star games?

## set up query
tbl = 'allstarfull'
query = 'SELECT * FROM ' + tbl

## execute
allstarfull = pd.read_sql_query(query, cnx)
print allstarfull.head()

## group by
allstar_grouped = allstarfull.groupby('playerid')

## find index of max
most = allstar_grouped.gamenum.count().argmax()

## df of player with most all star games
allstarfull[allstarfull.playerid == most]
```

```python
# Which school has generated the most expensive players? (expensive defined by their first year's salary).

query = """
    SELECT * 
        ,RANK() OVER(
            PARTITION BY playerid
                ORDER BY yearid ASC) AS year_in_league
    FROM salaries
"""

## execute
salaries_ranked = pd.read_sql_query(query, cnx)
print salaries_ranked.head()

## filter
salaries_ranked_yr1 = salaries_ranked[salaries_ranked.year_in_league == 1]
print salaries_ranked_yr1.head()

## join to schools
colleges_yr1salaries = salaries_ranked_yr1.merge()
```
