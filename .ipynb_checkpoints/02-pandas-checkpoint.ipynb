```python
print heart_data[heart_data['ca'].isnull()]
```

```python
""" Trial 1: Drop missing values. 
Pandas defines this on the axis (think axis of a matrix):
0: along the columns
1: along the rows
"""
print heart_data[285:290]

dropped_rows_heart_data = heart_data.dropna()
dropped_ca_thal_heart_data = heart_data.dropna(1)

print dropped_rows_heart_data[285:290]
print dropped_ca_thal_heart_data[285:290]

""" Trial 2: Fill the data
.ffill(): fills forward
.bfill(): fills backward
.fillna(): fills based on some value
"""

print heart_data[heart_data['ca'].isnull()]
print heart_data[285:290]['ca'].ffill()
print heart_data[285:290]['ca'].bfill()
print heart_data[285:290]['ca'].fillna(0) # fill with 0s
print heart_data[285:290]['ca'].fillna(int(heart_data['ca'].mode())) # fill with the most common value
```

```python
heart_data.describe()
```

```python
# thal (defects) is missing data as well
# categorical

heart_data[heart_data['thal'].isnull()]

"""
data cleaning options:
    set default to normal
    set as mode
    forward/backfill
"""

heart_data.thal.ffill()
heart_data.thal.bfill()

heart_data.thal.fillna(int(heart_data['thal'].mode()))
```

```python
import pandas as pd
import numpy as np
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'

header_row = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car name',]
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
mpg_data = pd.read_csv(url, header=0, na_values='?', delimiter='\s+')
mpg_data.columns = header_row # .columns attribute sets column names                     
```

```python
mpg_data.head()
```

```python
print mpg_data[mpg_data.model_year == 70].describe()
print mpg_data[mpg_data.model_year == 82].describe()

# '82 models are on average more fuel efficient than '70 models at the cost of horsepower
# oddly, '82 models have higher acceleration rates, possibly because they are ~1000 lbs lighter
```

```python
mpg_data[mpg_data.horsepower.isnull()]
```

```python
grouped = mpg_data.groupby('model_year')
grouped.horsepower.mean()
```

```python
years = [70, 82]
for year in years:
    print mpg_data[mpg_data['model_year'] == year].describe()
```

```python
for i in mpg_data['model_year'].unique():
    mpg_data[mpg_data['model_year'] == i].fillna(mpg_data[mpg_data['horsepower'].notnull()].mode(), inplace=True)    
```
