```python
# Which school has generated the most distinct players?

## Show all playerids and salaries with a salary in the year 1985 above 500k.

## set up query
tbl = 'collegeplaying'
query = 'SELECT * FROM ' + tbl

## execute
collegeplaying = pd.read_sql_query(query, cnx)
print collegeplaying.head()

## aggregate
collegeplaying_grouped = collegeplaying.groupby('schoolid')
print collegeplaying_grouped.playerid.nunique().argmax()
```

```python
# Which school has generated the most expensive players? (expensive defined by their first year's salary).

query = """
    SELECT * 
        ,RANK() OVER(
            PARTITION BY playerid
                ORDER BY yearid ASC) AS year_in_league
    FROM salaries
"""

## execute
salaries_ranked = pd.read_sql_query(query, cnx)
print salaries_ranked.head()

## filter
salaries_ranked_yr1 = salaries_ranked[salaries_ranked.year_in_league == 1]
print salaries_ranked_yr1.head()

## join to schools
colleges_yr1salaries = salaries_ranked_yr1.merge(collegeplaying, on='playerid')
print colleges_yr1salaries.head()

## aggregate
colleges_yr1salaries_grouped = colleges_yr1salaries.groupby('schoolid')
print colleges_yr1salaries_grouped.salary.sum().argmax()
```

```python
# Show the 5 most expensive salaries for each team in the year 2014.

query = """
    SELECT teamid
        ,salary
        ,salary_rank
    FROM
            (
                SELECT * 
                    ,ROW_NUMBER() OVER(
                        PARTITION BY teamid
                            ORDER BY salary DESC) AS salary_rank
                FROM salaries
                WHERE yearid = 2014
            ) sub
    WHERE salary_rank <= 5
    ORDER BY teamid, salary_rank
"""

## execute
topsalaries2014 = pd.read_sql_query(query, cnx)
topsalaries2014.head()
```

```python
## Partition the average salaries by team and year, against year. 
## Find players that were paid more than 1 standard deviation above the average salary for that team and year. 
## Show a count by playerid.

query = """
    SELECT *
        ,avg_yrteam + stddev_yrteam AS stddev_above_avg_yrteam
        ,CASE WHEN salary >= (avg_yrteam + stddev_yrteam) THEN 1 ELSE 0 END AS is_above
    FROM
        (
            SELECT yearid
                ,teamid
                ,playerid
                ,salary
                ,AVG(salary) OVER(PARTITION BY teamid, yearid) AS avg_yrteam
                ,STDDEV(salary) OVER(PARTITION BY teamid, yearid) AS stddev_yrteam
            FROM salaries
        ) sub
"""

## execute
salaries_above = pd.read_sql_query(query, cnx)
print salaries_above.head()

## aggregate
salaries_above_grouped = salaries_above.groupby(['yearid', 'teamid'])
salaries_above_grouped.is_above.sum()
```
