```python
from math import pi
from bokeh.plotting import *

df = pd.DataFrame(sp500)[:50] ## take first 50 observations from dataframe

mids = (df.Open + df.Close) / 2 ## find midpoints
spans = abs(df.Close - df.Open) ## find absolute difference in price from open to close

inc = df.Close > df.Open ## test whether close price was higher than open price, store results in list
dec = df.Open > df.Close ## test whether open price was higher than close price, store results in list
w = 12 * 60 * 60 * 1000 ## calculate 12 hours in milliseconds

output_notebook() # Load the Bokeh Java-Script interactive browser

figure(x_axis_type="datetime", width=1000, name="candlestick",
       tools="pan,wheel_zoom,box_zoom,reset,previewsave")

hold() 
## don't show the figure yet, there's more to do...

## defines the black bars
segment(df.index, df.High, df.index, df.Low, color='black')

## plot rectangles on days when stock price increases
## we're identifying the index on the dataframe for these days using the logical test that we stored in the variable "inc"
## specifically, we're identifying the index (the date)
rect(df.index[inc], mids[inc], w, spans[inc], fill_color="#D5E1DD", line_color="black")

## plot rectangles on days when stock price decreases
## we're identifying where on the dataframe these days occur using the logical test that we stored in the variable "inc"
## specifically, we're identifying the index (the date)
rect(df.index[dec], mids[dec], w, spans[dec], fill_color="#F2583E", line_color="black")

## rect("which data"=inc/dec, "center"=mids, "width"=w, "height"=spans, fill color, line color)

curplot().title = "SP500 Candlestick" ## add title
xaxis().major_label_orientation = pi/4 ## rotate labels
grid().grid_line_alpha = 0.3 ## change transparency

## ...ok now show it
show()
```

```python
import statsmodels.api as sm
import statsmodels.graphics.tsaplots as tsa

# Calculate Autocorrelation, Partial Auto Correlation and the Correlogram  
fig = plt.figure()
ax1 = fig.add_subplot(211)
fig = tsa.plot_acf(walk, lags = 500, ax = ax1)   
ax2 = fig.add_subplot(212)
fig = tsa.plot_pacf(walk, lags = 50, ax = ax2)
```

```python
N = 1000 # length of data
z = np.random.randn(N) # generate a list of random numbers to use as plot data
walk = np.zeros(N) # generate a list of 0's of the same length as z

for i in np.arange(1, N, 1):
    walk[i] = walk[i - 1] + z[i]
plt.plot(walk)

arma_mod22 = sm.tsa.ARMA(walk, (2, 2)).fit()
print(arma_mod22.summary())
print "Durbin Watson score: ", sm.stats.durbin_watson(arma_mod22.resid)
```

```python
from statsmodels.graphics.api import qqplot
from scipy import stats

resid = arma_mod22.resid
print stats.normaltest(resid)
fig = qqplot(resid, line = 'q', fit = True)
```

# Statsmodels ARMA Examples

```python
from __future__ import print_function
import numpy as np
import statsmodels.api as sm
import pandas as pd
from statsmodels.tsa.arima_process import arma_generate_sample
np.random.seed(12345)
```

```python
arparams = np.array([.75, -.25])
maparams = np.array([.65, .35])
```

```python
arparams = np.r_[1, -arparams]
maparam = np.r_[1, maparams]
nobs = 250
y = arma_generate_sample(arparams, maparams, nobs)
```

```python
dates = sm.tsa.datetools.dates_from_range('1980m1', length=nobs)
y = pd.TimeSeries(y, index=dates)
arma_mod = sm.tsa.ARMA(y, order=(2,2))
arma_res = arma_mod.fit(trend='nc', disp=-1)
```

```python
print(arma_res.summary())
```
