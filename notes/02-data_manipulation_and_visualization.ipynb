```python
# set up the environment

%pylab inline
import pandas as pd
import seaborn as sns
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'
```

```python
# load our data

bike_data = pd.read_csv('../DAT18NYC/data/bikeshare.csv')

print bike_data.head(24) # see what a day's worth of data looks like
bike_data.describe()
```

### Notes:
- hourly counts of active riders
- also broken down by rider type: "casual" riders are one-off riders, "registered" riders are subscribed
- includes some weather stats, looks like Farenheit in decimal (?)
- registered riders almost always outnumber casual riders

```python
# ridership by hour
bike_data_hourly = bike_data.groupby('hr')

# use sum to see total riders by hour
# threw in a sort to quickly see the busiests hours
bike_data_hourly.sum().sort('cnt', ascending=0)
```

### Notes:
- afternoon to early evening (12-8) clearly dominates
- morning (7-12) ridership is secondary
- there's a noticeable drop off during late night to early morning (10-7)

```python
# now, let's check out casual vs registered
print 'Casual:'
print bike_data_hourly['hr','casual'].sum().sort('casual', ascending=0).head()
print
print 'Registered:'
print bike_data_hourly['hr','registered'].sum().sort('registered', ascending=0).head()
```

### Notes:
- early afternoon is most popular with casuals
- definitely more of a rush hour presence for registered riders

```python
# ridership by day
bike_data_daily = bike_data.groupby('dteday')

# apply the same methodology to daily counts
bike_data_daily.sum().sort('cnt', ascending=0).head(20)
```

### Notes:
- September 2012 was a big month for the bikeshare, representing a large portion of the top dates
- There was also a solid leadup over the summer that carried over into October 2012

```python
# now, let's check out casual vs registered
print 'Casual:'
print bike_data_daily['hr','casual'].sum().sort('casual', ascending=0).head(10)
print
print 'Registered:'
print bike_data_daily['hr','registered'].sum().sort('registered', ascending=0).head(10)
```

### Notes:

So this is interesting:
- casual riders generated a lot of usage earlier in 2012 (Spring)
- registered riders show the most activity in September/October 2012 (the overall most active time period)

Curious to know if there's a reason for the distinction:
- did a lot of casual riders register later in the year?
- did something change in the pricing structure that made more people want to register?

Let's take a look at the impact that some of the other features of the data might have on ridership...

```python
# removing the lower-level time data
cols = ['season', 'mnth', 'temp', 'atemp', 'hum', 'windspeed', 'cnt']

# let's see what sticks
pd.scatter_matrix(bike_data[cols], figsize=(24, 24))
```

### Notes:
- No clear correlations with ridership counts here, but that's probably because the data's not in the right format.
- Let's try grouping by temperature since it correlates with both time-series data and some of the other weather data.
- There's an overrepresentation of January and December data, so keep that in mind

```python
temp_grp = bike_data[['temp', 'cnt']].groupby('temp').sum() # sum the counts by temperature
```

Plot it:

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(temp_grp.index, temp_grp.cnt, '.')
```

- Definitely looks to have some correlation.
- As the temperature goes up, ridership increases.  However, if the temperature gets too high, it will drop off.

Let's take a look at this data disaggregated by rider type...

```python
# unpivot by rider type and rollup the counts
melted = pd.melt(bike_data.reset_index(),
                 id_vars='temp',
                 value_vars=['casual','registered','cnt'],
                 value_name='sum')

temp_grp = melted.groupby(['temp','variable']).sum()
temp_grp = temp_grp.reset_index() # this is important! (otherwise the rplot won't recognize the column names)
```

```python
import pandas.tools.rplot as rplot

plt.figure(figsize=(10, 10), dpi=80)
plot = rplot.RPlot(temp_grp, x='temp', y='sum')
plot.add(rplot.TrellisGrid(['variable','.']))
plot.add(rplot.GeomScatter())
print plot.render(plt.gcf())
```

### Notes:
- Looks like registered riders are more likely to brave the elements.
- It's hard to see some of the nuance for the casual riders because of the y-axis (haven't learned that yet), but it does seem to lack the second "hump" around 40 degrees.  This gives some more support to the hypothesis that registered riders are more likely commuters.

# Summary
