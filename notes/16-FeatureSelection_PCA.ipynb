```python
import pandas as pd
from sklearn import feature_selection

def make_irisdf():
    from sklearn.datasets import load_iris
    from pandas import DataFrame
    iris = load_iris()
    df = DataFrame(iris.data, columns=iris.feature_names)
    df['target'] = iris.target
    return df

iris = make_irisdf()
```

```python
## variance threshold

## for each feature, print the data variance
print iris.ix[:,:4].apply(lambda x: x.var())

print iris.ix[:,:4].head()

## set the variance threshold to 0.6
## this only keeps the two features with variance above 0.6 (sepal length and petal length)
print feature_selection.VarianceThreshold(threshold=.6).fit_transform(iris.ix[:,:4])[:5]

## change threshold to .5
print feature_selection.VarianceThreshold(threshold=.5).fit_transform(iris.ix[:,:4])[:5]
```

```python
## select K best

print iris.ix[:,:4].head()
print

print 'F Test'
## create an instance of an f-test classifier that will select the 3 best features
ftest = feature_selection.SelectKBest(score_func=feature_selection.f_classif, k=3)

## run the data through the selector and print the scores
print pd.Series(ftest.fit(iris.ix[:,:4], iris['target']).scores_, index=iris.ix[:,:4].columns)

## transform the data, keeping the best 3 features (highest f-statistic)
print ftest.fit_transform(iris.ix[:,:4], iris['target'])[:5]
print

print 'Chi^2 Test'
## create an instance of a chi^2 test that will select the 3 best features
chi = feature_selection.SelectKBest(score_func=feature_selection.chi2, k=3)

## run the data through the selector and print the scores
print pd.Series(chi.fit(iris.ix[:,:4], iris['target']).scores_, index=iris.ix[:,:4].columns)

## transform the data, keeping the best 3 features (lowest chi^2)
print chi.fit_transform(iris.ix[:,:4], iris['target'])[:5]
```

```python
## regularization with logistic regression

from sklearn import linear_model as lm

## create an instance of a logistic regression using L1 and a error tolerance (cost) of 0.1
clf = lm.LogisticRegression(penalty='L1', C=0.1)

print iris.ix[:,:4].head()

## fit data and put coefficients into a dataframe
print pd.DataFrame(clf.fit(iris.ix[:,:4], iris['target']).coef_, columns=iris.ix[:,:4].columns)

## trasnform the data, if the majority of a feature's data is still predictive after regularization
print clf.fit_transform(iris.ix[:,:4], iris['target'])[:5]
```

```python
## tree
from sklearn import tree

## create and instance of a tree classifier
clf = tree.DecisionTreeClassifier(max_depth=4)
print iris.ix[:,:4].head()

## run the data through the classifier and print their importance
print pd.Series(clf.fit(iris.ix[:,:4], iris['target']).feature_importances_, index=iris.ix[:,:4].columns)

## transform data using the classifier
print clf.fit_transform(iris.ix[:,:4], iris['target'])[:5]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

random_data = pd.DataFrame({
        'x': range(1, 10),
        'y': range(1, 10),
        'z': range(1, 10),
    })

## covariance matrix
print np.cov(random_data.T, bias=1)

## variance of x
print np.var(random_data.x.T)
```

```python
## pull out eigenvalues
eig, Q = np.linalg.eig(np.cov(random_data.T, bias=1))

## print eigenvectors
print eig
print Q
```

```python
## amount of variance explained
## elbow rule

import seaborn as sns
from __future__ import division
eigsort = np.sort(eig)[::-1]
sns.set_style('white')
plt.figure()
plt.plot(range(1, len(eigsort) + 1), eigsort)
plt.xlabel('principal component')
plt.ylabel('explained variances')
plt.show()
```

```python
plt.figure()
plt.plot(range(1, len(eigsort) + 1), eigsort / max(eigsort))
plt.xlabel('principal component')
plt.ylabel('% explained variance')
plt.show()
```

```python
# ordering eigenvalues and vectors together
ordered = sorted(zip(eig, Q.T), reverse=True)
eig = np.array([_[0] for _ in ordered])
Q = np.column_stack((_[1] for _ in ordered))

# transforming data: We take the dot multiplication of the eigenvectors by the random data
X_transformed = np.dot(Q.T, random_data.T)
print X_transformed[0]
```

```python
plt.figure()
plt.plot(random_data.y, random_data.x, '.')
```

```python
plt.figure()
plt.plot(X_transformed[0], '.')
```

```python
from sklearn import decomposition
plt.plot(decomposition.PCA().fit_transform(random_data).T[0], '.')
```

```python
random_data_scattered = pd.DataFrame({
        'x': range(1, 100),
        'y': range(1, 100),
        'z': range(1, 100),
    })

random_data_scattered['y'] = random_data_scattered.y.apply(lambda y: y + np.random.normal(scale=10))
random_data_scattered['z'] = random_data_scattered.z.apply(lambda z: z + np.random.normal(scale=20))

print random_data_scattered.head()

plt.plot(random_data_scattered.x, random_data_scattered.y, '.')
```

```python
## pull out eigenvalues
eig2, Q2 = np.linalg.eig(np.cov(random_data_scattered.T, bias=1))

## print eigenvectors
print eig2
print Q2
```

```python
import seaborn as sns
from __future__ import division
eigsort2 = np.sort(eig2)[::-1]

sns.set_style('white')

plt.figure()
plt.plot(range(1, len(eigsort2) + 1), eigsort2 / max(eigsort2))
plt.xlabel('principal component')
plt.ylabel('% explained variance')
plt.show()
```
