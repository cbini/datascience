```python
import pandas as pd
import seaborn as sns
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'
```

```python
columns = """-- The attributes are (dontated by Riccardo Leardi, 
    riclea@anchem.unige.it )
    1) Alcohol
    2) Malic acid
    3) Ash
    4) Alcalinity of ash  
    5) Magnesium
    6) Total phenols
    7) Flavanoids
    8) Nonflavanoid phenols
    9) Proanthocyanins
    10)Color intensity
    11)Hue
    12)OD280/OD315 of diluted wines
    13)Proline"""

# we can generate the columns list using some string and list functions.
columns = columns.split('\n')
print columns
print
columns = [i.strip() for i in columns][2:]
print columns
print
columns = [i.split(')')[1].strip().lower() for i in columns]
print columns
print
cols = ['class']
cols.extend(columns)
print cols
print

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url, header=None)
wine.columns = cols
wine.head()
```

```python
# Create a groupby object:
winegroups = wine.groupby('class')
print type(winegroups)

# groupby objects allow for some standard pandas calls:

winegroups.describe()
```

```python
wine.groupby('class').ash.mean()
```

```python
# Groupby objects also:
## can work through selected columns

print winegroups.alcohol.mean()
print winegroups[['alcohol', 'ash']].median()

## can work through aggregation dictionaries, which can reference the origin
## groupby functions, or construct new functions.

print winegroups.agg({'alcohol': 'median'}) # agg function takes a dictionary
print winegroups.agg({'alcohol': lambda alcohol: alcohol.nunique() / alcohol.count()}) # percent of values in a column that are unique
```

```python
pivoted_wine = wine.pivot_table(values='ash', # values duh
                 columns='class', # pivot on
                 index='alcohol', # row
                 aggfunc='count') # aggregate

print pivoted_wine
```

```python
# group by advantages    
# pivot advantages

pivoted_wine_reset = pivoted_wine.reset_index()
print pivoted_wine_reset.head()
```

```python
# unpivot
print pd.melt(pivoted_wine_reset,
              id_vars=['alcohol'],
              value_vars=[1, 2, 3],
              value_name='count').dropna().head(20)
```
