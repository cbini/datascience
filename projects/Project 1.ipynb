```python
# get the data
stmath = pd.read_csv('../stmath.csv', index_col=0)

print len(stmath), 'observations', '\n'
stmath.head()
```

## Data definitions
| Feature | Type | Definition |
|-------|:----:|------------|
| **student_id** | Discrete | A unique identifer for each student |
| **school_name** | Discrete | A student's current school (academic year 2014-2015) |
| **grade_level** | Discrete | A student's current grade-level (academic year 2014-2015) |
| **cumulative_progress** | Continuous | The percentage completion that a student has achieved in their assigned ST Math grade-level module. If a student has completed a module and proceeded on to the next grade, it can be above 100. |
| **cumulative_mastery** | Continuous | Similiar to *cumulative_progress*, but inclusive of a student's "mastery" in their module's subject matter. If a student has completed a module and proceeded on to the next grade, it can be above 100. |
| **total_classroom_logins** | Continuous | The number of logins that a student has made at school |
| **total_home_logins** | Continuous | The number of logins that a student has made at home |
| **total_minutes logged** | Continuous | The total number of minutes a student has spent on the program, includes idle time |
| **SGP_MAP_Winter_Math (target variable)** | Continuous | A student's growth percentile from their baseline to winter score on the MAP math assessment. |
