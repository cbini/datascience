# Does ST Math usage have any impact on Student Growth Percentile?

"Blended Learning" and "Ed Tech" are two buzzwords that are becoming increasingly prominient in K-12 education.  Like in many other sectors, it seems like new startups or platforms are springing up every day, promising high student engagement and a personalized educational experience.  Educational apps now have thier own section of the annual SXSW tech conference, and some schools have gone as far as building their entire curriculum around them ([Rocketship](http://www.rsed.org/Blended-Learning.cfm), for example).

While, many educators are interested in how they can leverage these continuous developments in techonology, they are equal parts hesitant to integrate these programs into their lessons regularly. Teachers are uncertain about how effective these programs actually are, and it's difficult to choose which of the many programs will work the best for enriching a student's education.

This study will examine a new math program called [ST Math](http://www.mindresearch.org/programs/) and its impact, if any, on midyear growth measures from the math [MAP](https://www.nwea.org/assessments/map/) assessment: a nationally-normed, computer-adaptive assessment (à la the GRE) that our students take 2-3 times per year.  Looking at a number of usage statistics from ST Math, we will evalutate whether any of them having any impact on midyear Student Growth Percentile (SGP), a measure that shows how much a student has grown in a particular timeframe, relative to their peers across the country.

```python
## set up the environment
%pylab inline

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import sqlalchemy as sa
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'
```

```python
## create an database connection via sqlalchemy
engine = sa.create_engine('mssql+pyodbc://WINSQL01\\NARDO/KIPP_NJ')

## db query that will get our dataset
query = """
    WITH st_data AS (
      /*-- ST Math gives us cumulative data on a weekly basis.  This CTE summarizes YTD progress --*/
      SELECT student_number
            ,ROUND(SUM(K_5_Progress),0) AS cumulative_progress
            ,ROUND(SUM(K_5_mastery),0) AS cumulative_mastery
            ,SUM(num_lab_logins) + SUM(num_homework_logins) AS total_logins            
            ,SUM(minutes_logged) AS total_minutes_logged
      FROM
          (
           SELECT school_student_id AS student_number
                 ,week_ending_date
                 ,CASE
                   WHEN GCD = 'Kindergarten' THEN 0
                   WHEN GCD = 'First Grade' THEN 1
                   WHEN GCD = 'Second Grade' THEN 2
                   WHEN GCD = 'Third Grade' THEN 3
                   WHEN GCD = 'Fourth Grade' THEN 4
                   WHEN GCD = 'Fifth Grade' THEN 5
                   WHEN GCD = 'Sixth Grade' THEN 6
                   WHEN GCD LIKE '%Seven%' THEN 7
                  END AS GCD
                 ,K_5_Progress
                 ,K_5_mastery
                 ,num_lab_logins
                 ,num_homework_logins
                 ,SUM(minutes_logged_last_week) OVER(
                    PARTITION BY school_student_id, gcd) AS minutes_logged
                 ,ROW_NUMBER() OVER(
                    PARTITION BY school_student_id, gcd
                     ORDER BY week_ending_date DESC) AS rn -- highest achieved by grade-level module
           FROM STMath..progress_completion WITH(NOLOCK)     
           WHERE school_student_id IS NOT NULL
          ) sub
      WHERE rn = 1
      GROUP BY student_number
     )

    ,map_sgp AS (
      /*-- MAP mid-year SGP for math --*/
      SELECT studentid
            ,ROUND(CONVERT(FLOAT,growth_percentile),0) AS SGP_MAP_Winter_Math
      FROM KIPP_NJ..TABLEAU$MAP_SGP WITH(NOLOCK)
      WHERE year = 2014
        AND measurementscale = 'Mathematics'
        AND growth_term = 'MY'
        AND growth_percentile IS NOT NULL
     )

    SELECT ABS(CHECKSUM(NEWID())) AS student_id -- anonymous new id for each student
          ,co.school_name
          ,co.grade_level      
          ,st.cumulative_progress
          ,st.cumulative_mastery
          ,st.total_logins          
          ,st.total_minutes_logged      
          ,map.SGP_MAP_Winter_Math
    FROM KIPP_NJ..COHORT$identifiers_long#static co WITH(NOLOCK)
    JOIN st_data st
      ON co.student_number = st.student_number
    JOIN map_sgp map
      ON co.studentid = map.studentid
    WHERE co.year = 2014
      AND co.rn = 1
      AND co.schoolid != 999999
"""

## pandas <3
## getting data into a dataframe from a SQL database is as easy as feeding it a sqlalchemy engine and a query
stmath = pd.read_sql(sql=query, con=engine, index_col='student_id')

print len(stmath), 'observations'
stmath.head()
```

## SGP
SGP does not appear to be normal in this dataset, which is counterintuitive since most students should be clustered around the 50th percentile.

Using statsmodels and scipy, let's investigate SGP's normalcy with a QQ plot.

```python
sgp_middle = stmath_active[(stmath_active.SGP_MAP_Winter_Math < 75) & (stmath_active.SGP_MAP_Winter_Math >= 25)]

import statsmodels.api as sm
import scipy.stats as stats

print 'All Data:'
stmath_active.SGP_MAP_Winter_Math.hist()
fig = sm.qqplot(stmath_active.SGP_MAP_Winter_Math, dist=stats.distributions.norm, line='s')
plt.show()

print 'Middle Quartiles:'
sgp_middle.SGP_MAP_Winter_Math.hist()
fig = sm.qqplot(sgp_middle.SGP_MAP_Winter_Math, dist=stats.distributions.norm, line='s')
plt.show()

sgp_middle.describe()
```

So it actually isn't way off. Although, the tails are definitely pronounced.  

Correcting for them using the middle quartiles (the *sgp_middle* subset), the data much more closely resembles a normal distribution, so let's run with that moving forward.

## Conclusions and Next Steps

As it stands, there does not appear to be any meaningful correlation between MAP SGP and students' ST Math progress or thier usage of the system.  While it's disappointing that I wasn't able to lay the groundwork for a predictive model, I would definitely hesitate to call these definitive for a few reasons:

1. This is our district's first year using ST Math at a large scale.  Best practices and regular usage have not been adopted consistently across all schools and grade levels as they have been for the MAP assessment, so I imagine there would be a number of inconsistencies in these measures as well.  Additionally, the data quality that we receive from ST Math leaves a lot to be desired.  We only receive weekly aggregates of student progress from the ST Math

2. The MAP SGP distribution is puzzling, so I'm unsure as to whether plotting any of the features against it was valid.  As it is a percentile rank, one would expect most students to fall towards the middle. In fact, it was the opposite: students gravitated towards the extremes. The S-shape of it's Q-Q plot indicates that there might be some property of the data that I'm not picking up on.  Just as, I had to correct other features for their exponential nature, there may be some underlying aspect that needs to be accounted for as well.

3. This is preliminary data.  The norms for midyear SGP can be more of an end-of-year projection than a measure based on definitive midyear expected growth.  I'd definitely like to repeat this analysis at the end of year when we have a solid measure of student growth and more ST Math data.
